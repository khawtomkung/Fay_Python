### set operation

union
intersection
difference
complementary (not)

```python
set_a = {1, 2, 3}
set_b = {2, 3, 4}

print(set_a.intersection(set_b))

print(set_a | set_b | {4, 5, 6})

print(set_a.union([2, 3, 4]))

print(set_a.union(set_b).union({4, 5, 6}))
```

difference

```python
set_a = {1, 2, 3}
set_b = {2, 3, 4}

print(set_a.difference(set_b)) # A-B

print(set_a - set_b)

print(set_b.difference(set_a)) # B-A
```
