```python
numbers = []

for number in range(1, 101):
    numbers.append(number)

for number in numbers:
    if number % 3 == 0 and number % 2 != 0:
        numbers.remove(number)

print(len(numbers))
```

```python
numbers = []

for num in range(1, 101):
    if num % 3 != 0 or num % 2 == 0:
        numbers.append(num)

print(len(numbers))
```

```python
numbers = []

for num in range(1, 101):
    if num % 3 == 0 and num % 2 != 0:
        continue
    numbers.append(num)

print(len(numbers))
```
