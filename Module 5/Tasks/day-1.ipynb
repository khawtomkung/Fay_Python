## Recursive Function

```python
# !!! Beware of crashing !!!
def f(x):
    # Base Case
    if x > 100:
        return 0
    
    x += 1
    print(x)

    return f(x)

f(0)
# x = 1
# print(1)
# f(1)
# x = 2
# print(2)
# f(2)
# ...
# f(99)
# x = 100
# f(100)
# x = 101
# f(101)
```

```python
# 5! = 5 x 4 x 3 x 2 x 1
def recursive_factorial(x):
    if x == 1:
        return 1
    return x * recursive_factorial(x-1)

recursive_factorial(10)
```

```python
# 5! = 5 x 4 x 3 x 2 x 1
def interative_factorial(x):
    result = 1

    for number in range(1, x + 1):
        result *= number
    
    return result

interative_factorial(10)
```

```python
# %%timeit
import time

recursive_factorial(500)

t1 = time.time()
result = recursive_factorial(6)
t2 = time.time()
print('Time usage for recursive_factorial 6: ', t2-t1)
result = recursive_factorial(501)
t3 = time.time()
print('Time usage for recursive_factorial 500: ', t3-t2)
```

```python
# %%timeit
import time

interative_factorial(500)

t1 = time.time()
result = interative_factorial(6)
t2 = time.time()
print('Time usage for interative_factorial 6: ', t2-t1)
result = interative_factorial(501)
t3 = time.time()
print('Time usage for interative_factorial 500: ', t3-t2)
```

4. Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.

```python
# 4. Count uppercase and lowercase letters
def count_cases(string):
    upper = sum(1 for case in string if case.isupper())
    lower = sum(1 for case in string if case.islower())
    return upper, lower

count_cases('Hello Word')
```

6. Write a Python function to check whether a string is pangram or not.

Note : Pangrams are words or sentences containing every letter of the alphabet at least once. For example : "The quick brown fox jumps over the lazy dog"

```python
# 6. Check if string is pangram
def is_pangram(string):
    alphabet = set('abcdefghijklmnopqrstuvwxyz')
    if (alphabet.issubset(set(string.lower()))):
        return print('This is Pangram')
    else:
        return print('This is not Pangram')

is_pangram("The quick brown fox jumps over the lazy dog")
```

7. Write a Python function to calculate the fibonacci number of index i (starts with 1). The function accepts the number as an argument and return a single number. The fibonacci sequence starts with 1,1,2,3,5,8,13,21,.... Therefore, fibonacci of index 6 is number 8.

```python
# 7. Calculate fibonacci number at index i
def fibonacci(index):
    if index <= 0:
        return None
    if index == 1 or index == 2:
        return 1
    a, b = 1, 1
    for _ in range(3, index + 1):
        next_fib = a + b
        a = b
        b = next_fib
    return b

fibonacci(6)
```

8. Create a function which finds the length of each word in the phrase (broken by spaces) and return the values in a list. The function will have an input of a string, and output a list of integers.

9. Write a function that accepts a list of tuples containing the name of the student and their score, sort them by the score from high to low.

```python
# 9. Sort students by score (high to low)
def sort_by_score(student_list):
    return sorted(
        student_list,
        key=lambda student: student[1],
        reverse=True)

student_data = [
    ("Alice", 95),
    ("Bob", 88),
    ("Charlie", 100),
    ("David", 78),
    ("Eve", 95)
]
sort_by_score(student_data)
```

10. Write a python function that accepts a list of tuples of length 2, where each tuple contains student name and their scores respectively. The function should return the name of the student whose got maximum score. Please use map(), and max() function is strictly prohibited.

```python
# 10. Find student with maximum score using map()
def top_student(student_scores):
    scores_tuples = list(map(lambda item: (item[1], item[0]), student_scores))
    sorted_scores = sorted(scores_tuples, reverse=True)
    top_score = sorted_scores[0]
    return top_score[1]
        
data_list = [
    ("Alice", 95),
    ("Bob", 88),
    ("Charlie", 100),
    ("David", 78),
    ("Fiona", 99)
]

top_student(data_list)
```
