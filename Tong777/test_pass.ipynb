```python
def clear_screen():
    os.system('clear' if os.name != 'nt' else 'cls')

def press_to_continue():
    print("\nPress any key to continue...", end="", flush=True)
    fd = sys.stdin.fileno()
    old_settings = termios.tcgetattr(fd)
    try:
        tty.setraw(fd)
        sys.stdin.read(1)
    finally:
        termios.tcsetattr(fd, termios.TCSADRAIN, old_settings)
    print()

def loading_screen():
    symbols = ["‚ô†Ô∏é ‚ô§ ‚ô£Ô∏é ‚ôß", "‚ô§ ‚ô£Ô∏é ‚ôß ‚ô•Ô∏é", "‚ô£Ô∏é ‚ôß ‚ô•Ô∏é ‚ô°", "‚ôß ‚ô•Ô∏é ‚ô° ‚ô¶Ô∏é", "‚ô•Ô∏é ‚ô° ‚ô¶Ô∏é ‚ô¢", "‚ô° ‚ô¶Ô∏é ‚ô¢ ‚ô†", "‚ô¶Ô∏é ‚ô¢ ‚ô†Ô∏é ‚ô§", "‚ô¢ ‚ô†Ô∏é ‚ô§ ‚ô£Ô∏é"]
    time.sleep(1)
    for symbol in symbols:
        clear_screen()
        print(tong_777_pic)
        print(loading_pic)
        print(f"\n\n                                                     {symbol}\n")
        time.sleep(0.55)
    print(tong_777_pic)
    clear_screen()
    time.sleep(1)
```

```python
BASE_PATH = "/Users/kung/Module4/Fay_Python/Tong777/ID_user"
os.makedirs(BASE_PATH, exist_ok=True)

def load_player(username):
    filename = os.path.join(BASE_PATH, f"{username}.txt")
    if not os.path.exists(filename):
        return None
    with open(filename, "r") as file:
        data = file.read().strip().split(",")
        if len(data) != 3:
            return None
        name, hashed_password, money = data
        return hashed_password, float(money)

def save_player(username, hashed_password, money):
    filename = os.path.join(BASE_PATH, f"{username}.txt")
    with open(filename, "w") as file:
        file.write(f"{username},{hashed_password},{money:.2f}")
```

```python
def login_or_register():
    while True:
        clear_screen()
        print(tong_777_pic)
        print(login_pic)

        username = input("Enter username: ").strip()
        player_data = load_player(username)

        if player_data is not None:
            hashed_password, money = player_data
            password = input("Enter your password: ").encode('utf-8')

            if bcrypt.checkpw(password, hashed_password.encode('utf-8')):
                print(f"Welcome back, {username}!")
                press_to_continue()
                return username, hashed_password, money
            else:
                print("‚ùå Incorrect password.")
                press_to_continue()
        else:
            choice = input("User not found. Create a new account? (y/n): ").lower()
            if choice == 'y':
                password = input("Set your password: ").encode('utf-8')
                hashed_password = bcrypt.hashpw(password, bcrypt.gensalt()).decode('utf-8')
                money = 100.00
                save_player(username, hashed_password, money)
                print(f"‚úÖ Account '{username}' created successfully!")
                print("üéÅ You received 100.00 credits as a new player bonus üéÅ")
                press_to_continue()
                return username, hashed_password, money
            elif choice == 'n':
                continue
            else:
                print("Please type only 'y' or 'n'.")
                press_to_continue()
```

```python
def game_high_low(money):
    print("üéØ High-Low Game üéØ")
    bet = float(input(f"You have {money:.2f}. Enter your bet: "))
    if bet <= 0 or bet > money:
        print("Invalid bet amount!")
        return 0.0

    num1 = random.randint(1, 100)
    print(f"First number: {num1}")
    guess = input("Will the next number be (h)igher or (l)ower? ").lower()
    num2 = random.randint(1, 100)
    print(f"Next number: {num2}")

    if (guess == 'h' and num2 > num1) or (guess == 'l' and num2 < num1):
        print(f"üéâ You won {bet:.2f}!")
        return bet
    else:
        print(f"üíÄ You lost {bet:.2f}.")
        return -bet

def game_guess_number(money):
    print("üî¢ Guess the Number (1-10) üî¢")
    bet = float(input(f"You have {money:.2f}. Enter your bet: "))
    if bet <= 0 or bet > money:
        print("Invalid bet amount!")
        return 0.0

    correct = random.randint(1, 10)
    guess = int(input("Guess a number between 1 and 10: "))

    if guess == correct:
        print(f"üéØ Correct! The number was {correct}.")
        return bet
    else:
        print(f"‚ùå Wrong! The correct number was {correct}.")
        return -bet

def game_coin_flip(money):
    print("ü™ô Coin Flip (h/t) ü™ô")
    bet = float(input(f"You have {money:.2f}. Enter your bet: "))
    if bet <= 0 or bet > money:
        print("Invalid bet amount!")
        return 0.0

    guess = input("Choose (h)eads or (t)ails: ").lower()
    result = random.choice(['h', 't'])

    if guess == result:
        print(f"üéâ You won! It was {'heads' if result == 'h' else 'tails'}.")
        return bet
    else:
        print(f"üíÄ You lost! It was {'heads' if result == 'h' else 'tails'}.")
```

```python
def main():
    loading_screen()

    while True:
        username, hashed_password, money = login_or_register()

        while True:
            clear_screen()
            print(tong_777_pic)
            print(f"Player: {username}")
            print(f"Balance: {money:.2f} credits\n")
            print("1. Play High-Low")
            print("2. Play Guess Number")
            print("3. Play Coin Flip")
            print("4. Logout")
            print("5. Exit\n")

            choice = input("Select menu (1-5): ")

            if choice == '1':
                money += game_high_low(money)
                save_player(username, hashed_password, money)
                press_to_continue()

            elif choice == '2':
                money += game_guess_number(money)
                save_player(username, hashed_password, money)
                press_to_continue()

            elif choice == '3':
                money += game_coin_flip(money)
                save_player(username, hashed_password, money)
                press_to_continue()

            elif choice == '4':
                print("üîÅ Logging out...")
                time.sleep(1)
                break  # Go back to login/register

            elif choice == '5':
                print("üíæ Saving data and exiting game...")
                save_player(username, hashed_password, money)
                time.sleep(1)
                sys.exit()

            else:
                print("Please select 1-5 only.")
                press_to_continue()
```

```python

if __name__ == "__main__":
    main()
```
